```python
y_ad.value_counts() ## as we can see the outcome is well balanced
```

```python
model = svm.SVC(kernel='poly',degree=6,class_weight='balanced', probability=True) ## support vectory classify , it will divide the data in equally linear part
## i have used poly kernel as the values are more non linear , and increased the degree for better accuracy as it can be more flexible
```

```python
model.fit(x_train,y_train) ## it will fit the data and divide
```

## Parkinsons Disease Prediction Model

```python
Parkinsons_d = pd.read_csv('./Datasets/parkinsons.csv')
```

```python
Parkinsons_d.tail()
```

```python
## as there is a string data type column so will be removing it
extra_string = Parkinsons_d['name']
Parkinsons_d = Parkinsons_d.drop(columns=['name'])
```

```python
Parkinsons_d['status'].value_counts()
## data is very imbalanced
```

```python
Parkinsons_d.groupby('status').mean()
```

```python
Parkinsons_d.describe()
```

```python
a = Parkinsons_d.drop(columns=['status'],axis=1)
b = Parkinsons_d['status']
```

```python
a.head()
```

```python
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size=0.2,random_state=49)
```

```python
a_train.head()
```

```python
## to make the data balance lets try using ADASYN method
## as we have already created adasyn before lets use it again in diabetes part
a_ad,b_ad = adasyn.fit_resample(a_train,b_train)
```

```python
b_ad.value_counts() ## as we can see the dataset is balanced now
```

```python
model3 = svm.SVC(kernel='poly',degree=3).fit(a_ad,b_ad)
```

```python
b_pred = model3.predict(a_test)
```

```python
print(accuracy_score(b_test,b_pred)) ## accuracy is good using degree 3
```
